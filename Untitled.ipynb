```python
data = pd.read_csv('Dimuon_DoubleMu.csv')

pd.set_option('display.max_columns', None)

data.head(100)
```

```python
    def approximateMass(num):
        tempMass = int(num*1000)
        if tempMass%100<=50:
            tempMass = tempMass - tempMass%100
            return float(tempMass/1000)
        if tempMass%100<=100:
            tempMass = tempMass - tempMass%100 + 100
            return float(tempMass/1000)
```

```python
masses = data.M
masses = sorted(masses)
for i in range(0, len(masses)-1):
    masses[i] = approximateMass(masses[i])
    print(masses[i])
```

```python
plt.figure(figsize = (15, 5))
plt.hist(masses, 1000, color='black', label='Mass', range = (0,102))
plt.title('Muon Mass')
plt.show()
```

```python
counts = np.array([])
massesSet = np.array([])
a=0
while a<len(masses):
    massesSet = np.append(massesSet, masses[a])
    counts = np.append(counts, 1)
    a = a+1
    while masses[a-1]==masses[a%(len(masses)-1)]:
        counts[-1] = counts[-1] + 1
        a = a+1
```

```python
for k in range(0, len(massesSet)-1):
    print(str(massesSet[k]) + ' occurs ' + str(counts[k]))
```
